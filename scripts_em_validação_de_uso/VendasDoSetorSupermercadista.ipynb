```python
import requests
import urllib
import json
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
url = r'https://api.bcb.gov.br/dados/serie/bcdata.sgs.7414/dados?formato=json'
response = urllib.request.urlopen(url)

data = json.loads(response.read())

data = pd.DataFrame(data)[100:300]
```

```python
plt.plot(data.data, data.valor)

plt.show()
```
